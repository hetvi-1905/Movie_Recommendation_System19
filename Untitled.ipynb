```python
import ast  # to convert string list into list
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head(1)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head(2)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3 :
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(convert3)
```

```python
movies.head(3)
```

```python
movies["crew"][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head(3)
```

```python
movies["overview"][0]
```

```python
# converting string format of overview column into list
movies["overview"] = movies["overview"].apply(lambda x : x.split()) 
# it will have comma after every word , space is default delimiter of method split so after each space the content will be seperated by comma
```

```python
movies.head(3)
```

```python
# removing spaces between two words of name like [ Sam Worthington  ---->  SamWorthington ]
```

```python
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head(3)
```

```python
movies["tags"] =  movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
# stemming
# ["loved,loving,lovely"] ------------> o/p: [love,love,love]
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():                  #------------------> converting string to list (with commas)
        y.append(ps.stem(i))   
    
    return " ".join(y)                             # ---------------->converting list to string again (no commas betweenS)
```

```python
ps.stem("loved")
```

```python
ps.stem("loving")
```

```python
ps.stem("lovely")
```

```python
stem('loving loved lovely')
```

```python
new_df["tags"] =  new_df["tags"].apply(stem)
```

```python
cv.get_feature_names() 
```

```python
# to again run cv for seeing changes after stemming
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
vectors = cv.fit_transform(new_df["tags"]).toarray()
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape # similarity of 4806 movies with 4806 movies is a martix of nXn where n = 4806
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(similarity[0],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"]==movie].index[0]
    similarity_distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(similarity_distances)),reverse=True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
# taking out index 
print(new_df[new_df["title"]=="Spectre"].index[0])
print(new_df[new_df["title"]=="Avatar"].index[0])
```

```python
recommend("Batman")
```
